## Section 4: Sentiment Analysis (VADER)
Applying VADER sentiment analysis on video titles to classify them as positive, negative, or neutral, and examining sentiment patterns across countries, categories, and engagement levels.

```python
import sys
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.polarity_scores(df_all['title'][0])
```

```python
df_all['sentiment_score'] = df_all['title'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
df_all[['title', 'sentiment_score']].head()
```

```python
def get_sentiment_label(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df_all['sentiment_label'] = df_all['sentiment_score'].apply(get_sentiment_label)
```

```python
df_all[['title', 'sentiment_score', 'sentiment_label']].head()
df_all['tags_sentiment_score'] = df_all['tags'].fillna('').apply(lambda x: analyzer.polarity_scores(x)['compound'])
df_all['tags_sentiment_label'] = df_all['tags_sentiment_score'].apply(get_sentiment_label)

df_all[['tags', 'tags_sentiment_score', 'tags_sentiment_label']].head()
```

### Overall Sentiment Distribution

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=df_all, x='sentiment_label', hue='sentiment_label', palette='Set2', legend=False, order=['Positive', 'Neutral', 'Negative'])
plt.title('Overall Sentiment Distribution of Video Titles')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Sentiment Distribution by Country

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_all, x='country', hue='sentiment_label', palette='Set2', order=['US', 'GB', 'IN', 'JP', 'DE'])
plt.title('Sentiment Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

### Average Views by Sentiment

```python
avg_views_sentiment = df_all.groupby('sentiment_label')['views'].mean().reindex(['Positive', 'Neutral', 'Negative'])

plt.figure(figsize=(7, 5))
sns.barplot(x=avg_views_sentiment.index, y=avg_views_sentiment.values, hue=avg_views_sentiment.index, palette='Set2', legend=False)
plt.title('Average Views by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Average Views')
plt.tight_layout()
plt.show()
```

### Sentiment Distribution by Top 6 Categories

```python
top_cats = df_all['category_id'].value_counts().head(6).index
df_top_cats = df_all[df_all['category_id'].isin(top_cats)]

category_map = {1: 'Film & Animation', 10: 'Music', 22: 'People & Blogs', 23: 'Comedy', 24: 'Entertainment', 25: 'News & Politics'}
df_top_cats = df_top_cats.copy()
df_top_cats['category_name'] = df_top_cats['category_id'].astype(int).map(category_map)

plt.figure(figsize=(12, 6))
sns.countplot(data=df_top_cats, x='category_name', hue='sentiment_label', palette='Set2', order=list(category_map.values()))
plt.title('Sentiment Distribution by Top 6 Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

## Section 5: Time Series Analysis
Examining how trending activity varies over time, by day of week, and by month, using the cleaned and merged dataset spanning November 14, 2017 to June 14, 2018.

### Trending Videos Over Time by Country

```python
trend_by_date = df_all.groupby(['trending_date', 'country']).size().reset_index(name='count')

plt.figure(figsize=(14, 6))
sns.lineplot(data=trend_by_date, x='trending_date', y='count', hue='country')
plt.title('Trending Videos Over Time by Country')
plt.xlabel('Date')
plt.ylabel('Number of Trending Videos')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

### Trending Videos by Day of Week

```python
df_all['day_of_week'] = df_all['trending_date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

day_counts = df_all['day_of_week'].value_counts().reindex(day_order)

plt.figure(figsize=(10, 6))
sns.barplot(x=day_counts.index, y=day_counts.values, hue=day_counts.index, palette='Set2', legend=False)
plt.title('Trending Videos by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Trending Videos by Month

```python
df_all['year_month'] = df_all['trending_date'].dt.to_period('M').astype(str)

monthly_counts = df_all.groupby('year_month').size()

plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_counts.index, y=monthly_counts.values, marker='o', color='steelblue')
plt.title('Trending Videos by Month')
plt.xlabel('Month')
plt.ylabel('Number of Trending Videos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
